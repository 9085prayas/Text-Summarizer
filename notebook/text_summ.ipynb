```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "google/pegasus-xsum"

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=False)

pegasus_model = AutoModelForSeq2SeqLM.from_pretrained(
    model_name,
    use_safetensors=True   # ✅ use this to avoid .bin loading
).to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
import pandas as pd
from datasets import DatasetDict, Dataset

train_df = pd.read_csv('samsum-train.csv')
test_df = pd.read_csv('samsum-test.csv')
validation_df = pd.read_csv('samsum-validation.csv')

ds = DatasetDict({
    'train': Dataset.from_pandas(train_df),
    'test': Dataset.from_pandas(test_df),
    'validation': Dataset.from_pandas(validation_df)
})
```

```python
split_length = [len(ds[split]) for split in ds]
print(f'Split lengths: {split_length}')
print(f'Features: {ds["train"].column_names}')

print('\nDilogues :')

print(ds['test'][1]['dialogue'])

print('\nSummary :')

print(ds['test'][1]['summary'])
```

```python
sample_dialogues = ds['train']['dialogue'][:2]
sample_summaries = ds['train']['summary'][:2]

print("Sample Dialogues:", sample_dialogues)
print("Sample Summaries:", sample_summaries)

try:
    input_encodings = tokenizer(sample_dialogues, max_length=1024, truncation=True, return_tensors="pt")
    target_encodings = tokenizer(text_target=sample_summaries, max_length=128, truncation=True, return_tensors="pt")

    print("\nInput Encodings:", input_encodings)
    print("\nTarget Encodings:", target_encodings)

except TypeError as e: #error handeling
    print(f"\nTypeError during direct tokenization: {e}")
except Exception as e:
    print(f"\nAn unexpected error occurred during direct tokenization: {e}")
```

```python
import torch

def convert_example_to_feature(example_batch):
    # Tokenize inputs
    # Set padding_side to 'left' for the encoder input
    tokenizer.padding_side = 'left'
    input_encodings = tokenizer(
        example_batch['dialogue'],
        max_length=200,
        truncation=True,
        padding='max_length',
        return_tensors='pt' # Return as PyTorch tensors
    )

    # Tokenize targets
    # Keep padding_side as default ('right') for the decoder input
    tokenizer.padding_side = 'right'
    with tokenizer.as_target_tokenizer():
      target_encodings = tokenizer(
        text_target=example_batch['summary'],
        max_length=128,
        truncation=True,
        padding='max_length',
        return_tensors='pt' # Return as PyTorch tensors
    )

    # Replace padding token IDs in labels with -100
    labels = [
        [(token if token != tokenizer.pad_token_id else -100) for token in label_seq]
        for label_seq in target_encodings["input_ids"]
    ]

    # Convert labels to torch.LongTensor and move to device
    labels_tensor = torch.LongTensor(labels)


    return {
        'input_ids': input_encodings['input_ids'],
        'attention_mask': input_encodings['attention_mask'],
        'labels': labels_tensor # Use the tensor here
    }
```

```python
ds_pt = ds.filter(lambda example: 'dialogue' in example and example['dialogue'] and 'summary' in example and example['summary']) #data cleaning checking if the dialogue or summary is not empty
ds_pt = ds_pt.map(convert_example_to_feature, batched=True)
```

```python
ds_pt['train']
```

```python
#Training
#this cell takes data as patches (batched data input)
from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer,
                                               model=pegasus_model,
                                               label_pad_token_id=-100)
```

```python
from transformers import Trainer , TrainingArguments

training_args = TrainingArguments(
      output_dir='pegasus-samsum', warmup_steps=500,
      per_device_train_batch_size=1,
      gradient_accumulation_steps=4,
      num_train_epochs=1,
      learning_rate=2e-4,
      fp16=True,
      save_total_limit=3,
      logging_steps=1,
      optim="adamw_torch", # Changed optimizer
      lr_scheduler_type="cosine",
      warmup_ratio=0.05,
)
```

```python
trainer = Trainer(model=pegasus_model, args=training_args,
                  tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                  train_dataset=ds_pt['train'],
                  eval_dataset=ds_pt['validation'])
```

```python
print(f"Model vocabulary size: {pegasus_model.config.vocab_size}")
print(f"Tokenizer vocabulary size: {tokenizer.vocab_size}")
```

```python
trainer.train()
```

```python
# prompt: save the model

#save model
model_path = "./pegasus-samsum-model"
tokenizer_path = "./tokenizer"

trainer.model.save_pretrained(model_path)
tokenizer.save_pretrained(tokenizer_path)
```
